# Full 24h

```python
import os
import pandas as pd
import numpy as np

# Directory paths
base_dir = 'data/psykose'  # Folder containing class folders
output_dir = 'data_24h/psykose'  # Folder to store the processed data
os.makedirs(output_dir, exist_ok=True)

# Define function to split time series into 24-hour chunks
def split_time_series(file_path, segment_duration=24):
    """
    Splits a time series CSV file into 24-hour segments.
    
    Parameters:
    - file_path: str, path to the CSV file
    - segment_duration: int, duration of each segment (in hours)
    
    Returns:
    - List of DataFrames, each representing a segment
    """
    df = pd.read_csv(file_path)
    if 'timestamp' not in df.columns:
        raise ValueError("The CSV file must have a 'timestamp' column.")
    
    # Ensure timestamp is in datetime format and sorted
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df = df.sort_values('timestamp')
    
    # Calculate 24-hour chunks (assuming the data is hourly sampled)
    start_time = df['timestamp'].min()
    end_time = df['timestamp'].max()
    total_hours = int((end_time - start_time).total_seconds() // 3600)
    
    segments = []
    for i in range(0, total_hours, segment_duration):
        segment = df[
            (df['timestamp'] >= start_time + pd.Timedelta(hours=i)) &
            (df['timestamp'] < start_time + pd.Timedelta(hours=i + segment_duration))
        ]
        if not segment.empty:
            segments.append(segment)
    
    return segments

# Process all class folders
for class_folder in os.listdir(base_dir):
    class_path = os.path.join(base_dir, class_folder)
    if not os.path.isdir(class_path):
        continue  # Skip non-folder entries
    
    print(f"Processing class: {class_folder}")
    output_class_dir = os.path.join(output_dir, class_folder)
    os.makedirs(output_class_dir, exist_ok=True)
    
    for file_name in os.listdir(class_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(class_path, file_name)
            segments = split_time_series(file_path)
            
            # Save each segment as a new CSV file
            for idx, segment in enumerate(segments):
                segment_file = os.path.join(output_class_dir, f"{file_name[:-4]}_segment_{idx}.csv")
                segment.to_csv(segment_file, index=False)
                print(f"Saved: {segment_file}")
```

```python
import os
import pandas as pd
from collections import defaultdict

def count_rows_in_csv_files(folder_path):
    # Tworzymy defaultdict, aby automatycznie tworzy listy dla kadej liczby wierszy
    row_counts_dict = defaultdict(int)
    
    # Iterujemy przez pliki w folderze
    for filename in os.listdir(folder_path):
        # Sprawdzamy, czy plik ma rozszerzenie .csv
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            
            # Wczytujemy plik CSV
            try:
                df = pd.read_csv(file_path)
                row_count = len(df)
                # Zwikszamy licznik dla tej liczby wierszy
                row_counts_dict[row_count] += 1
            except Exception as e:
                # Obsuguje ewentualne bdy podczas otwierania pliku
                print(f"Error reading {filename}: {e}")
    
    return row_counts_dict

# cieka do folderu z plikami CSV
folder_path = './data_24h/psykose/condition'


# Wywoanie funkcji i wywietlenie wynikw
row_counts_grouped = count_rows_in_csv_files(folder_path)

# Wypisanie wynikw w wymaganej formie
for row_count, file_count in row_counts_grouped.items():
    print(f"Dugo {row_count}: {file_count} plikw")
```

```python
import os
import pandas as pd
import numpy as np

# Directory paths
base_dir = 'data/depresjon'  # Folder containing class folders
output_dir = 'data_day/depresjon'  # Folder to store the processed data
os.makedirs(output_dir, exist_ok=True)

# Define function to split time series into 13-hour chunks (from 8:00 AM to 9:00 PM)
def split_time_series(file_path, segment_start_hour=8, segment_duration=13):
    """
    Splits a time series CSV file into 13-hour chunks from 8:00 AM to 9:00 PM.
    
    Parameters:
    - file_path: str, path to the CSV file
    - segment_start_hour: int, the starting hour for each segment (default is 8 for 8:00 AM)
    - segment_duration: int, duration of each segment (in hours)
    
    Returns:
    - List of DataFrames, each representing a segment
    """
    df = pd.read_csv(file_path)
    if 'timestamp' not in df.columns:
        raise ValueError("The CSV file must have a 'timestamp' column.")
    
    # Ensure timestamp is in datetime format and sorted
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df = df.sort_values('timestamp')
    
    # Calculate segments based on the specified start hour and duration
    segments = []
    start_time = df['timestamp'].min()
    end_time = df['timestamp'].max()

    # Loop over each day in the data and create a 13-hour segment from 8:00 AM to 9:00 PM
    current_time = start_time.replace(hour=segment_start_hour, minute=0, second=0)
    while current_time + pd.Timedelta(hours=segment_duration) <= end_time:
        # Get the segment from 8:00 AM to 9:00 PM
        segment = df[
            (df['timestamp'] >= current_time) & 
            (df['timestamp'] < current_time + pd.Timedelta(hours=segment_duration))
        ]
        if not segment.empty:
            segments.append(segment)
        
        # Move to the next day (next 8:00 AM)
        current_time += pd.Timedelta(days=1)
    
    return segments

# Process all class folders
for class_folder in os.listdir(base_dir):
    class_path = os.path.join(base_dir, class_folder)
    if not os.path.isdir(class_path):
        continue  # Skip non-folder entries
    
    print(f"Processing class: {class_folder}")
    output_class_dir = os.path.join(output_dir, class_folder)
    os.makedirs(output_class_dir, exist_ok=True)
    
    for file_name in os.listdir(class_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(class_path, file_name)
            segments = split_time_series(file_path)
            
            # Save each segment as a new CSV file
            for idx, segment in enumerate(segments):
                segment_file = os.path.join(output_class_dir, f"{file_name[:-4]}_segment_{idx}.csv")
                segment.to_csv(segment_file, index=False)
                print(f"Saved: {segment_file}")
```
